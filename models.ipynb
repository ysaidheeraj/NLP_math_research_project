```python
import pandas as pd
data = pd.read_csv('/content/data_with_ner_pos_vocabcount.csv')
```

```python
data.columns
```

```python
data.drop(columns = ['CARDINAL', 'DATE', 'EVENT', 'FAC', 'GPE', 'LANGUAGE', 'LAW', 'LOC',
       'MONEY', 'NORP', 'ORDINAL', 'ORG', 'PRODUCT', 'WORK_OF_ART'], inplace = True)
```

```python
data.drop(columns = ['problem', 'modified_problem'], inplace = True)
```

```python
data.head()
```

```python
for i in range(data.shape[0]):
  data['level'][i] = data['level'][i][6]
```

```python
data.head()
```

```python
X = data.drop(columns = ['level'])
y = data['level']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2, stratify = y)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

def evaluation(y_train_pred, y_test_pred, y_train, y_test):
  print("train_evaluation:\n")
  print(classification_report(y_train_pred, y_train))
  print(confusion_matrix(y_train_pred, y_train))
  print("\ntest_evaluation:\n")
  print(classification_report(y_test_pred, y_test))
  print(confusion_matrix(y_test_pred, y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(class_weight='balanced')
dt.fit(X_train, y_train)
y_test_pred = dt.predict(X_test)
y_train_pred = dt.predict(X_train)
evaluation(y_train_pred, y_test_pred, y_train, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression(class_weight='balanced',max_iter=1000)
```

```python
log.fit(X_train, y_train)
y_test_pred = log.predict(X_test)
y_train_pred = log.predict(X_train)
evaluation(y_train_pred, y_test_pred, y_train, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=25,class_weight='balanced')
```

```python
rfc.fit(X_train, y_train)
y_test_pred = rfc.predict(X_test)
y_train_pred = rfc.predict(X_train)
evaluation(y_train_pred, y_test_pred, y_train, y_test)
```
